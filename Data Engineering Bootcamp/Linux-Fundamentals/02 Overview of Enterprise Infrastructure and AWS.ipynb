### Overview of Hardware
Let us understand few details related to Hardware before we deep dive into the material.

The Essential Components of computer
* **Memory** – Good to have 8GB or 16GB RAM.
* **CPU** – Processor can be dual core or quad core. Quad-core has double the speed than dual-core.16GB with quad-core is an ideal combination.
* **Storage** – Can be SSD or Hard disk. Storage doesn’t affect the performance. Data can be stored on the cloud as well.
* **Networking** – Better internet connectivity will ensure good performance.

### Personal Computer vs. Enterprise Server
Any Computer is nothing but a combination of CPU, Memory, Storage and then necessary software. However, there are subtle differences between Personal Computer and Enterprise Server. Let us go through those differences.
* Enterprise Infrastructure is built to support mission critical applications.
* Storage required for Enterprise Application Components such as Databases is separated from servers.
* There will multiple redundant network cards connected with multiple cables. There will be different network switches for private and public communication.
* Our personal computers are designed typically to run a user-friendly operating system and desktop applications to facilitate desktop-oriented tasks.

### Overview of Data Centers
Enterprises set up servers in either their data centers or rent servers from Data Centers of existing infrastructure providers. Let us understand briefly about a Data Center.
* A data center is a facility composed of networked computers, network switches, and storage servers to support mission-critical applications of an enterprise.
* Typical components of a data center include servers, storage hardware, cables, and racks, as well as a variety of information security elements such as firewalls.
* Data centers contain a set of routers and switches that transport traffic between the servers and to the outside world.

### Physical Machine vs. Virtual Machine
As Hardware configurations evolved, Enterprise Infrastructure such as Servers, Network Switches, Storage becomes extremely powerful. It is not practical to have one application running on one physical server. To use resources more efficiently Virtualization came into the picture. Let us understand the differences between Physical Machine vs. Virtual Machine briefly.
* The physical machine itself contains all the resources such as CPU, Memory, Storage etc and each of them will take some real estate.
* These are cost ineffective when compared to ec2 or virtual machines, where we can create several Virtual Machines from one Physical Machine. We can isolate applications and run on different servers with limited Real Estate.
* We can easily manage them as resources are pooled and the Virtualization or Cloud provide High Availability and Fault Tolerance of the Virtual Machines.
* EC2 or a Virtual Machine running on a data center which built using physical machines.

### Signing up to AWS
AWS stands for Amazon Web Services. It is leading cloud platform where we can quickly provision Virtual Machines. Let us first understand how to sign up to the platform.
* Creating an account is very simple just with a working email id is required.
* It will ask for your credit card information, but you will not be charged.
* You can get the alerts while any charges on your account.
* An ec2(elastic compute cloud) is a virtual machine built on Aws data center.
* There are different regions you have to handle them carefully.
* You can use one of the quick start AMI(Amazon Machine Image), a better option is the free tier.
* After selecting you will see the pricing, focus and be aware of the charge.
* AWS marketplace is for any software engineer to create AMI images or templates.
* An instance is free to use for 750 hours per month along 12 months.
* Select all the options appropriately and launch after review.
* Select an existing key pair and download it.
* Secure the key pair even with the permissions to it.
* Click on launch instances, it will create a virtual machine.

### Provisioning EC2 Instance from AWS
As part of this session, we will see how to provision instance from AWS. We will actually go through some of the key concepts toward the end of this module.
* Go to the Web Console and then EC2 Dashboard.
* Click on Launch Instance and search for OS of your choice (Centos 7 in our case)
* Choose instance type
* Configure Network
* Add any additional storage you want to mount on to the instance
* Configure Security Groups
* Launch instance by either creating new key pair or by using an existing key pair.

### Connecting to EC2 Instance
Once we provision Virtual Machine or EC2 Instance from AWS, we might have to connect to use it for deploying applications. As part of this topic, we will see how we can connect to EC2 instance using SSH.

Remove if extensions are with the downloaded pem file(key pair) and copy it to your preferred location.

After copying make sure to check and if needed change the permissions on the pem file. eg <mark>chmod 600 ~/pemfile</mark>

If you are using windows you can use Cygwin, putty or windows subsystem for Linux or any ssh client to connect ec2 instance using the Pem file(key pair). It is good to setup windows subsystem for Linux or Cygwin on your system to connect any Linux based remote machine.

You can use either public IP or public DNS to connect ec2 instance with pem file authentication. Click on connect to get the username for the instance to connect.

### Basic Linux Commands
Let us run some basic linux commands on EC2 instance just to get our selves used to the environment.
* <mark>df -h</mark> is to check the storage of the machine.
* <mark>lscpu</mark> is to check CPU count on the machine.
* <mark>free -h</mark> to check memory details.
* <mark>sudo</mark>(superuser do) is to execute a command with admin privileges.
* <mark>uname -a</mark>. to get the operating system details.
* <mark>sudo su - root</mark> To access root account from a normal user account.
